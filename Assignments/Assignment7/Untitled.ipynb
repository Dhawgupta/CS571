```python
n_in, n_h, n_out, batch_size = 2,5,1,10 # first we will train a XOR gate
# generate random data for the xor gate
n = 10000 # ddataset size
x = [[random.randint(0,1),random.randint(0,1)] for i in range(n)]
y = [i^j for i,j in x]

```

```python
for i,k in enumerate(x): # seeing the validity of the dataset produced
    print(x[i],y[i])
```

```python
# now we will train the nueral netword
X = torch.tensor(x,dtype = torch.float32)
Y = torch.tensor(y, dtype = torch.float32)
Y = Y.reshape((-1,1))
```

```python
X = torch.Tensor([[0,0],[0,1],[1,0],[1,1]])
Y = torch.Tensor([0,1,1,0]).view(-1,1)

```

```python
print(X)
print(Y)
```

```python
model = nn.Sequential(nn.Linear(n_in, n_h), nn.Sigmoid(), nn.Linear(n_h, n_out))
```

```python
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
for epoch in range(10000):
    # forward propoagete
    y_pred = model(X)
    
    # computer the loss
    loss = criterion(y_pred, Y)
    print("Epoch : ", epoch, ' Loss : ', loss.item())
    
    # zero teh gradients
    optimizer.zero_grad()
    
    # perform the backprop
    loss.backward()
    
    # update the parametrs
    optimizer.step()
    
```

```python
X = torch.Tensor([[0,0],[0,1],[1,0],[1,1]])
Y = torch.Tensor([0,1,1,0]).view(-1,1)
```

```python
import torch.nn.functional as F
class XOR(nn.Module):
    def __init__(self, input_dim = 2, output_dim = 1):
        super(XOR, self).__init__()
        self.lin1 = nn.Linear(input_dim, 2)
        self.lin2 = nn.Linear(2, output_dim)
        self.rel1 = nn.ReLU()
        self.rel2 = nn.ReLU()
    def forward(self, x):
        x = self.lin1(x)
#         x = self.rel1(x)
        x = F.sigmoid(x)
        x = self.lin2(x)
#         x = F.sigmoid(x)
        return x
```

```python
model = XOR()
loss_func = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.02, momentum=0.1)
```

```python
from torch.autograd import Variable
epochs = 10000
steps = X.size(0)
for i in range(epochs):
    for j in range(steps):
        data_point = np.random.randint(X.size(0))
        x_var = Variable(X[data_point], requires_grad=False)
        y_var = Variable(Y[data_point], requires_grad=False)
        
        optimizer.zero_grad()
        y_hat = model(x_var)
        loss = loss_func.forward(y_hat, y_var)
        loss.backward()
        optimizer.step()
        
    if i % 500 == 0:
        print ("Epoch: {}, Loss: {}, ".format(i, loss.data.numpy()))
```
