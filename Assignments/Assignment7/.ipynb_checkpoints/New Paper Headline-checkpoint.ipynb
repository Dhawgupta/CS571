```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import numpy as np
import random
import pandas as pd
```

```python
# first get the data
```

```python

my_data = np.genfromtxt('AllDataset.csv', delimiter=',')
```

```python
# remove the first col becuse it is just the column nubmer
data = my_data[1:]
```

```python
# now we need to extract the X and Y values
y = data[:,-1]
x = data[:,:-1]

y = to_categorical(y)
print(x.shape)
print(y.shape)
```

```python
model = Sequential()
model.add(Dense(units = 100, activation='relu',input_dim = x.shape[1]))
model.add(Dense(units = 5, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=100, batch_size=32)
```

```python
loss_and_metrics = model.evaluate(x, y, batch_size=128)
print("Loss : {}\nAccuracy : {}".format(loss_and_metrics[0], loss_and_metrics[1]))
```

```python
y = data[:,-1]
x = data[:,:-1]

y = to_categorical(y)
print(x.shape)
print(y.shape)
model3 = Sequential()
model3.add(Dense(units = 1000, activation='relu',input_dim = x.shape[1]))
model3.add(Dense(units = 500, activation='relu'))
model3.add(Dense(units = 100, activation='relu'))
model3.add(Dense(units = 5, activation = 'softmax'))
model3.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
model3.fit(x, y, epochs=100, batch_size=32)
```

```python
loss_and_metrics3 = model3.evaluate(x, y, batch_size=128)
print(loss_and_metrics3)
```

```python
# Do this with a train test split: 
seed = 7
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=seed)
model4 = Sequential()
model4.add(Dense(units = 1000, activation='relu',input_dim = x.shape[1]))
model4.add(Dense(units = 500, activation='relu'))
model4.add(Dense(units = 100, activation='relu'))
model4.add(Dense(units = 20, activation='relu'))
model4.add(Dense(units = 5, activation = 'softmax'))
model4.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
model4.fit(X_train, y_train, epochs=120, batch_size=32)
loss_and_metrics = model4.evaluate(X_test, y_test, batch_size=128)
print(loss_and_metrics)
```
